```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

LAT = 19.0760
LON = 72.8777
API_KEY = "1573876b7b9629f4ec85517706ad8177"

url = f"https://api.openweathermap.org/data/2.5/forecast?lat={LAT}&lon={LON}&appid={API_KEY}&units=metric"
response = requests.get(url)
data = response.json()

# Extract forecast data
dates = []
temps = []
humidity = []
wind_speed = []

for item in data['list']:
    dates.append(item['dt_txt'])
    temps.append(item['main']['temp'])
    humidity.append(item['main']['humidity'])
    wind_speed.append(item['wind']['speed'])

# Create DataFrame
weather_data = pd.DataFrame({
    "Date": dates,
    "Temperature (°C)": temps,
    "Humidity (%)": humidity,
    "Wind Speed (m/s)": wind_speed
})

print(weather_data.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample Weather Data
data = {
    "Date & Time": ["2025-02-25 12:00:00", "2025-02-25 15:00:00", "2025-02-25 18:00:00"],
    "Temperature (°C)": [22.00, 24.17, 20.50],
    "Humidity (%)": [60, 55, 70],
    "Pressure (hPa)": [1012, 1010, 1013],
    "Wind Speed (m/s)": [4.5, 5.2, 3.8],
    "Weather Description": ["Clear Sky", "Few Clouds", "Scattered Clouds"]
}

# Convert to DataFrame
df = pd.DataFrame(data)
df["Date & Time"] = pd.to_datetime(df["Date & Time"])  # Convert to datetime

# Add 'Feels Like Temperature' (Simple Approximation)
df["Feels Like (°C)"] = df["Temperature (°C)"] + (df["Humidity (%)"] * 0.05) - (df["Wind Speed (m/s)"] * 0.1)

# Data Cleaning (Check & Handle Missing Values)
df = df.dropna()  # Drop missing values if any

# Visualization: Temperature Trends
plt.figure(figsize=(8, 5))
sns.lineplot(x=df["Date & Time"], y=df["Temperature (°C)"], marker='o', label="Actual Temperature")
sns.lineplot(x=df["Date & Time"], y=df["Feels Like (°C)"], marker='s', label="Feels Like Temperature")
plt.xlabel("Date & Time")
plt.ylabel("Temperature (°C)")
plt.title("Temperature Trend")
plt.legend()
plt.xticks(rotation=45)
plt.show()

# Visualization: Humidity & Pressure Trends
fig, ax1 = plt.subplots(figsize=(8, 5))
ax2 = ax1.twinx()
ax1.plot(df["Date & Time"], df["Humidity (%)"], 'g-^', label="Humidity (%)")
ax2.plot(df["Date & Time"], df["Pressure (hPa)"], 'b-o', label="Pressure (hPa)")
ax1.set_xlabel("Date & Time")
ax1.set_ylabel("Humidity (%)", color='g')
ax2.set_ylabel("Pressure (hPa)", color='b')
plt.title("Humidity vs. Pressure")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create DataFrame
df = pd.DataFrame(data)
df["Date & Time"] = pd.to_datetime(df["Date & Time"])

# Rolling mean (window = 2)
df["Temp Rolling Mean"] = df["Temperature (°C)"].rolling(window=2).mean()

plt.figure(figsize=(8,5))
plt.plot(df["Date & Time"], df["Temperature (°C)"], marker='o', label="Actual Temp")
plt.plot(df["Date & Time"], df["Temp Rolling Mean"], linestyle='--', label="Rolling Mean")
plt.title("Temperature Trend with Rolling Average")
plt.xlabel("Date & Time")
plt.ylabel("Temperature (°C)")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_df = df[["Temperature (°C)", "Humidity (%)", "Pressure (hPa)", "Wind Speed (m/s)"]]

corr = numeric_df.corr()

plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Weather Parameters")
plt.show()
```

```python
Q1 = df["Temperature (°C)"].quantile(0.25)
Q3 = df["Temperature (°C)"].quantile(0.75)
IQR = Q3 - Q1

outliers = df[
    (df["Temperature (°C)"] < Q1 - 1.5 * IQR) |
    (df["Temperature (°C)"] > Q3 + 1.5 * IQR)
]

print("Detected Outliers:")
print(outliers)
```

```python
weather_freq = df["Weather Description"].value_counts()

plt.figure(figsize=(6,4))
weather_freq.plot(kind="bar")
plt.title("Weather Condition Frequency")
plt.xlabel("Weather Type")
plt.ylabel("Count")
plt.xticks(rotation=30)
plt.show()
```

```python
# Create Heat Index–like feature
df["Heat Index"] = (
    df["Temperature (°C)"] * 0.7 +
    df["Humidity (%)"] * 0.2 -
    df["Wind Speed (m/s)"] * 0.1
)

plt.figure(figsize=(6,5))
plt.scatter(df["Humidity (%)"], df["Heat Index"])
plt.xlabel("Humidity (%)")
plt.ylabel("Heat Index")
plt.title("Humidity vs Heat Index Relationship")
plt.grid(True)
plt.show()
```
